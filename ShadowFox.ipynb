```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
```

```python
data={'ID number':[6725,6734,6739,6740,6751,6753,6754],
      'NAME':['Shiva','Vijay','Surya','Mahender','Hussain','Rudra','Akshath'],
      'SGPA':[9.15,9.90,9.90,9.20,9.02,8.75,9.20],
      'CGPA':[9.38,9.70,9.47,9.11,9.11,8.89,9.35]
      }
df=pd.DataFrame(data)
df=df.sort_values(by='CGPA')
df=df.reset_index(drop=True)
df
```

LINE CHART

```python
##SeaBorn
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['ID number'], y=df['CGPA'], marker='o', color='blue')
plt.xlabel('ID number')
plt.ylabel('CGPA')
plt.title('CGPA vs ID number (SeaBorn)')
plt

##Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(df['ID number'], df['CGPA'], marker='o', color='blue')
plt.xlabel('ID number')
plt.ylabel('CGPA')
plt.title('CGPA vs ID number (Matplotlib)')
plt
```

HISTOGRAM

```python
##SeaBorn
plt.figure(figsize=(10, 6))
sns.histplot(df['SGPA'], bins=6, kde=True, color='purple')
plt.xlabel('SGPA')
plt.ylabel('Frequency')
plt.title('Distribution of SGPA (Seaborn)')
plt

##Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df['SGPA'], bins=6, color='purple', edgecolor='black')
plt.xlabel('SGPA')
plt.ylabel('Frequency')
plt.title('Distribution of SGPA (Matplotlib)')
plt
```

BAR GRAPH

```python
##SeaBorn
plt.figure(figsize=(10,6))
sns.barplot(x=df['NAME'], y=df['CGPA'], palette='Blues_d')
plt.xlabel('Name')
plt.ylabel('CGPA')
plt.title('CGPA vs Name (Seaborn)')
plt.xticks(rotation=90)
plt

##Matplotlib
plt.figure(figsize=(10,6))
plt.bar(df['NAME'], df['CGPA'], color='blue', edgecolor='black')
plt.xlabel('Name')
plt.ylabel('CGPA')
plt.title('CGPA vs Name (Matplotlib)')
plt.xticks(rotation=90)
plt
```
